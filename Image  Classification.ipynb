#### Project Overview
This project focuses on **image classification** for three categories of images: **Fauna**, **Footwear**, and **Vehicles**. The goal is to classify images into these categories using machine learning models.



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from sklearn.utils import shuffle
from skimage.transform import resize
```

```python
import os
import glob

# Print the current working directory
print("Current Working Directory:", os.getcwd())

# Check if the path exists
path = 'Dataset/Fauna'
if os.path.exists(path):
    print("Path exists!")
else:
    print("Path does NOT exist!")

# Check for files in the folder
print("Files in the folder:", os.listdir(path) if os.path.exists(path) else "Folder not found!")

# Use glob to find `.jpg` files
fauna = glob.glob(os.path.join(path, "*.jpeg"))
print("Number of images found:", len(fauna))
print("Image paths:", fauna)
```

```python
print("Current Working Directory:", os.getcwd())

path = 'Dataset/Footwear'
if os.path.exists(path):
    print("Path exists!")
else:
    print("Path does NOT exist!")

print("Files in the folder:", os.listdir(path) if os.path.exists(path) else "Folder not found!")

footwear = glob.glob(os.path.join(path, "*.jpeg"))
print("Number of images found:", len(footwear))
print("Image paths:", footwear)
```

```python
print("Current Working Directory:", os.getcwd())

path = 'Dataset/Vehicles'
if os.path.exists(path):
    print("Path exists!")
else:
    print("Path does NOT exist!")

print("Files in the folder:", os.listdir(path) if os.path.exists(path) else "Folder not found!")

vehicle = glob.glob(os.path.join(path, "*.jpeg"))
print("Number of images found:", len(vehicle))
print("Image paths:", vehicle)
```

```python
fauna = glob.glob('Dataset/Fauna/*.*')
footwear = glob.glob('Dataset/Footwear/*.*')
vehicle = glob.glob('Dataset/Vehicles/*.*')
```

```python
image = []
label = []

for i in fauna:
    img = plt.imread(i)
    image.append(img)
    label.append(0)
    
for i in footwear:
    img = plt.imread(i)
    image.append(img)
    label.append(1)
    
for i in vehicle:
    img = plt.imread(i)
    image.append(img)
    label.append(2)
    
```

```python
print(f"Number of images loaded: {len(image)}")
```

```python
plt.figure(figsize = (10,8))
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

```python
image, label = shuffle(image, label)
```

```python
plt.figure(figsize = (10,8))
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(image[i])
plt.tight_layout()
```

```python
for i in range(10):
    print(image[i].shape)
```

```python
X = []

for i in image:
    X.append(resize(i, (500,500,3)))
```

```python
X = np.array(X)
```

```python
y = np.array(label)
```

```python
d1, d2, d3, d4 = X.shape
```

```python
X = X.reshape((d1, d2*d3*d4))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

##### SVM Model

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
%time model.fit(X_train, y_train)
```

```python
%time y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot = True)
```

```python
print(classification_report(y_test,y_pred))
```
